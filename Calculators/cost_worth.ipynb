```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rel_worth = np.array([0,1,2,3,4,5,10,15,20,25,30,35,40,45,50])
lower = np.array([0,0,0,0,0,0,0,11,17,23,28,34,39,44,49])
upper = np.array([10,10,10,10,11,11,14,18,22,27,32,36,41,46,51])

components = {
    "Toolbox Shell": (32, 46),  # (x=Worth, y=Cost)
    "Work Surface": (26, 23),
    "Tool Tray": (16, 0),
    "Legs": (18,14),
    "Feet": (8,3)
}

```

```python
plt.figure(figsize=(10, 6))

# Plotting the bounds
plt.plot(rel_worth, lower, label='Lower Bound', color='blue', lw=2)
plt.plot(rel_worth, upper, label='Upper Bound', color='red', lw=2)



# Formatting
plt.title('Cost-Worth Analysis', fontsize=14)
plt.xlabel('Relative Worth', fontsize=12)
plt.ylabel('Cost', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()

for name, (x, y) in components.items():
    # Plot the point
    plt.scatter(x, y, color='black', zorder=5) 
    
    # Add the label
    plt.annotate(
        name, 
        xy=(x, y),            # The point to annotate
        xytext=(8, 8),        # Offset text by 8 points (x, y)
        textcoords='offset points', 
        fontsize=10,
        fontweight='bold',
        arrowprops=dict(arrowstyle='->', color='black', lw=0.5) # Optional arrow
    )
# Display
plt.tight_layout()
plt.show()
```
